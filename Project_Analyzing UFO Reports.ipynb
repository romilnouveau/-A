```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
%matplotlib inline  
```

```python
ufo_data=pd.read_csv(r'C:\Users\shreya\Desktop\scrubbed.csv', low_memory=False)
ufo_data['latitude'] = pd.to_numeric(ufo_data['latitude'],errors='coerce')
ufo_data['duration (seconds)'] = pd.to_numeric(ufo_data['duration (seconds)'],errors='coerce')
```

# Countries that are the part of the dataset

```python
us_data=ufo_data[(ufo_data['country']=='us')]
ca_data=ufo_data[(ufo_data['country']=='ca')]
gb_data=ufo_data[(ufo_data['country']=='gb')]
au_data=ufo_data[(ufo_data['country']=='au')]
de_data=ufo_data[(ufo_data['country']=='de')]
us=plt.scatter(us_data['longitude'],us_data['latitude'],c='c',s=10)
ca=plt.scatter(ca_data['longitude'],ca_data['latitude'],c='coral',s=20)
gb=plt.scatter(gb_data['longitude'],gb_data['latitude'],c='b',s=20)
au=plt.scatter(au_data['longitude'],au_data['latitude'],c='lightblue',s=20)
de=plt.scatter(de_data['longitude'],de_data['latitude'],c='y',s=20)
plt.legend((us,ca,gb,au,de),
           ('US', 'CANADA', 'G.BRITAIN', 'AUSTRALIA', 'GERMANY'))
plt.xlabel('longitudes')
plt.ylabel('latitudes')
plt.show()
```

# UFO Sightings in different countries

```python
shape_plot=sns.countplot(x='country',data=ufo_data)
shape_plot.set_xticklabels(shape_plot.get_xticklabels())
```

```python
print('Number of cases by percentage')
ct=np.array([len(us_data), len(ca_data), len(gb_data), len(au_data), len(de_data)])
ctt=np.array(['us','ca','gb','au','de'])
for i in range(0,5):
    print(ctt[i],':  ',ct[i]/len(ufo_data)*100,'%')
```

# Different UFO shapes reported

```python
shape_plot=sns.countplot(x='shape',data=ufo_data)
shape_plot.set_xticklabels(shape_plot.get_xticklabels(), rotation=90)
plt.title('Different UFO shapes')
```

```python
print('The percentage of different UFO shapes reported')
ss=np.array(ufo_data['shape'].unique())
for i in range(0,30):
    print(ss[i],': ',len(ufo_data[(ufo_data['shape']==ss[i])])/len(ufo_data)*100,'%')
```

# Reports in different states of the country

```python
fig, axes = plt.subplots(figsize=(15,3))
us_state_plot=sns.countplot(x='state',data=us_data, order = us_data['state'].value_counts().index)
us_state_plot.set_xticklabels(us_state_plot.get_xticklabels())
plt.title("No. of UFO's reported in United States")
```

**California(ca) has so far reported the maximum number(>8000) of UFO Sightings.**

```python
fig, axes = plt.subplots(figsize=(15,3))
plt.subplot(1,3,1)
ca_state_plot=sns.countplot(x='state',data=ca_data, order = ca_data['state'].value_counts().index)
ca_state_plot.set_xticklabels(ca_state_plot.get_xticklabels())
plt.title("No. of UFO's reported in Canada")
plt.subplot(1,3,2)
gb_state_plot=sns.countplot(x='state',data=gb_data)
gb_state_plot.set_xticklabels(gb_state_plot.get_xticklabels())
plt.title("No. of UFO's reported in Great Britain")
plt.subplot(1,3,3)
au_state_plot=sns.countplot(x='state',data=au_data, order = au_data['state'].value_counts().index)
au_state_plot.set_xticklabels(au_state_plot.get_xticklabels())
plt.title("No. of UFO's reported in Australia")
```

**Ontario(on) in canada has reported the maximum no.(>1250) of UFO reports.**

The data of GB & AU is missing because of the null values in the state column.

```python
from datetime import datetime
ufo_data['datetime'] = pd.to_datetime(ufo_data['datetime'],errors='coerce')
ufo_data.insert(1,'year',ufo_data['datetime'].dt.year)
ufo_data.insert(2,'month',ufo_data['datetime'].dt.month)
ufo_data.insert(3,'time',ufo_data['datetime'].dt.time)
```

```python
ufo_data.head() #separated date,time and year of the reports
```

```python
t=ufo_data[(ufo_data['year']>2013.0)]['year'].apply(lambda x:x-100)
s=ufo_data[(ufo_data['year']<=2013.0)]['year']
```

# UFO Reports by year

```python
a=pd.DataFrame({'year':t}) 
b=pd.DataFrame({'year':s}).sort_values('year')
yy=pd.concat([a,b]).sort_values('year') 
fig, axes = plt.subplots(figsize=(20,5))
plot=sns.countplot(x='year',data=yy)
plot.set_xticklabels(plot.get_xticklabels(),rotation=90)
plt.title("UFO Reports by year")
```

**The number of UFO Sightings has increased significantly with advancing era.**

# **UFO Reports during different times of the year**

```python
fig, axes = plt.subplots(figsize=(15,10))
plt.subplot(2,2,1)
us_months=ufo_data[(ufo_data['country']=='us')]
usplot=sns.countplot(x='month',data=us_months)
usplot.set_xticklabels(usplot.get_xticklabels())
plt.title("US")
plt.subplot(2,2,2)
ca_months=ufo_data[(ufo_data['country']=='ca')]
caplot=sns.countplot(x='month',data=ca_months)
caplot.set_xticklabels(caplot.get_xticklabels())
plt.title("Canada"
plt.subplot(2,2,3)
gb_months=ufo_data[(ufo_data['country']=='gb')]
gbplot=sns.countplot(x='month',data=gb_months)
gbplot.set_xticklabels(gbplot.get_xticklabels())
plt.title("Great Britain")
plt.subplot(2,2,4)
au_months=ufo_data[(ufo_data['country']=='au')]
auplot=sns.countplot(x='month',data=au_months)
auplot.set_xticklabels(auplot.get_xticklabels())
plt.title("Australia")
```

**Different Countries have reported maximum UFOs during different times of the year.**

While US has reported a good number of cases during the later half of the year(maximum in July), Canada, on the other hand, has faced a substantial drop in the UFO reports except in the middle of the year(i.e. July & August)

The trend of UFO Reports in Great Britain is very similar to that of US. Australia has the maximum number of cases not only in the middle of the year(June) but also in the beginning of the year i.e. January.

Overall, the sightings have been done mostly during **summers.**
